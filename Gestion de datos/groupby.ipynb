```python
import pandas as pd
df = pd.read_csv("C:/Users/luigi/data_science/db/base_datos_2008.csv")
```

```python
# groupby() agrupa por el parametro by

# Agrupación por día de la semana de acuerdo al retraso del vuelo
# Obtener el día con más retraso en el vuelo
df.groupby(by="DayOfWeek")["ArrDelay"].max()
```

```python
# Obtener el día con el minimo retraso en el vuelo
df.groupby(by="DayOfWeek")["ArrDelay"].min()
```

```python
# Obtener la medía del grupo de vuelo (día de la semana) segun el retraso de vuelo

df.groupby("DayOfWeek")["ArrDelay"].mean()
```

```python
# Obtener cantidad de vuelos por semana
# Cantidad de elementos en cada grupo

df.groupby("DayOfWeek")["ArrDelay"].count()
```

```python
# Resumen estadistico

df.groupby("DayOfWeek")["ArrDelay"].describe()
```

```python
puntuales = df[df["ArrDelay"] == 0]
```

```python
puntuales["ArrDelay"]
```

```python
# Obteniendo porcentaje de puntualidad
total = len(puntuales["ArrDelay"])/ len(df)
```

```python
print('Total de vuelos: ', len(df))
print('Vuelos que llegan a tiempo: ', len(puntuales))
print('Porcentaje de puntualidad: ' + str(round(total * 100, 2)) + '%')
```
