```python
import pandas as pd
df = pd.read_csv("C:/Users/luigi/data_science/db/base_datos_2008.csv", nrows = 1e6)
```

```python
# Duplicando casos (filas)

dfduplicate = pd.concat([df, df])
```

```python
# Salteando datos
dfduplicate = dfduplicate.sample(frac=1)
```

```python
# Eliminando casos duplicados
dfclean = dfduplicate.drop_duplicates()
```

```python
len(dfclean) == len(df)
```

```python
# Buscar duplicados sobre una columna

dfclean.drop_duplicates(subset="DayofMonth")
```
