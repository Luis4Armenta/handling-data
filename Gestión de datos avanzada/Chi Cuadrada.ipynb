```python
import pandas as pd
import numpy as np
df = pd.read_csv("C:/Users/luigi/data_science/db/base_datos_2008.csv")
```

$$ {x^2} = \sum \limits _{i=j} ^k \frac{({observando _i} - {esperado _i}) ^2 }{esperado _i} $$

```python
np.random.seed(0)
df = df[df['Origin'].isin(['HOU', 'ATL', 'IND'])]
df = df.sample(frac=1)
df = df[0:10000]
```

```python
df['BigDelay'] = df['ArrDelay'] > 30
observados = pd.crosstab(index=df['BigDelay'], columns=df['Origin'], margins=True)
```

```python
observados
```

```python
round(len(df[(df['Origin'] == 'ATL') & (df['BigDelay'] == False)]) / len(df[df['Origin'] == 'ATL' ]), 3) * 100
```

```python
from scipy.stats import chi2_contingency
```

```python
test = chi2_contingency(observados)
```

```python
# 1. Suma de las diferencias al cuadrado
# 2. p valor (permite tomar desiciones basado en un criterio)
# Tabla de valores esperados

test
```

```python
esperados = pd.DataFrame(test[3])
```

```python
esperados
```

```python
esperados_rel = round(esperados.apply(lambda r: r/len(df) *100, axis=1),2)
```

```python
observados_rel = round(observados.apply(lambda r: r/len(df) *100, axis=1),2)
```

```python
observados_rel
```

```python
esperados_rel
```

```python
test[1]
```

## Resumen de Test de hipótesis
    * Si el p-valor < 0.05 hay diferencias significativas: Hay relación entre variables
    * si el p-valor > 0.05 no hay diferencias significativas: No hay relación entre variables
