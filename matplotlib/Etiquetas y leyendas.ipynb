```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(0)

df = pd.read_csv('C:/Users/luigi/data_science/db/base_datos_2008.csv', nrows=100000)
data = np.unique(df.DayOfWeek, return_counts=True)
labs = ['Lun', 'Mar', 'Mie', 'jue', 'Vie','SÃ¡b', 'Dom']
data
```

```python
plt.pie(
         x=data[1],
         #labels=labs,
         radius=.7,
         colors=sns.color_palette('hls', 7),
         startangle=90,
         explode=(0,0,0,0,0,0,0.1),
         autopct="%1.1f%%",
         labeldistance=1.2
        )
plt.legend(loc='best', labels=labs)
plt.show()
```
