```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
np.random.seed(0)

df = pd.read_csv('C:/Users/luigi/data_science/db/base_datos_2008.csv', nrows=1000000)
df = df.sample(frac=1).head(100)
```

```python
plt.scatter(x=df.DayofMonth, y=df.ArrDelay, s=df.Distance)
```

```python
plt.scatter(x=df.DayofMonth, y=df.ArrDelay, s=df.Distance, alpha=.3)
```

```python
plt.scatter(x=df.DayofMonth, y=df.ArrDelay, s=df.Distance, alpha=.3, c=df.DayOfWeek.isin([6,7]))
plt.title("Retrasos en EEUU")
plt.xlabel("Díá del mes")
plt.ylabel("Retraso al lugar")
plt.ylim([0, 150]) #sum a la grafica
plt.xticks([0, 15, 30])
plt.text(x=28, y=128,s="mi vuelo")
```
